# Scatter plot 3
    Con las siguientes características:
    label = "años"
    color = verde
    simbolo * ---> marker = '*'
    s = 500 es bastante grande, tamaño del punto

```python
plt.scatter(x,y,label = 'años', color = 'green', marker = '*', s = 500)
plt.title('Mi scatter plot!')
plt.xlabel(' Eje X')
plt.ylabel(' Eje Y')
plt.legend()
plt.show()
```

# Pie chart (gráficos circulares)

```python
y = ["Futbol", "Baloncesto", "Gimnasio", "TV"]
x = [4, 2, 4, 1]
```

```python
plt.pie(x, labels = y)
plt.title('Horas semanales')
plt.show()
```

```python
plt.pie(x, labels = y, 
       colors = ['r','y','b','g'],# colores
       startangle = 90,# angulo de inicio
       radius = 1)# radio
plt.title('Horas semanales')
plt.show()
```

```python
plt.pie(x, labels = y, 
       colors = ['r','y','b','g'],
       startangle = 90,
       radius = 1,
       autopct = '%1.2f%%')# porcentaje con dos decimales
plt.title('Horas semanales')
plt.show()
```

# Varios plots

```python
plt.figure(figsize = (12,8))

x = np.arange(-10,10,0.01)


plt.subplot(2, 2, 1); y1 = x ; plt.plot(x, y1) ; plt.title("y = x")

plt.subplot(2, 2, 2); y2 = x**2 ; plt.plot(x, y2); plt.title("y es x al cuadrado")

plt.subplot(2, 2, 3); y3 = x**3 ; plt.plot(x, y3); plt.title("y es x al cubo")

plt.subplot(2, 2, 4); y4 = x**4 ; plt.plot(x, y4); plt.title("y es x a la cuarta")

plt.show()
```
