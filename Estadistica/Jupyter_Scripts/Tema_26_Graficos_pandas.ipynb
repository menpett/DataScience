```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({"X": [10, 20, 30, 40, 50], "Y":[15, 5, 10, 8, 6]})
```

```python
df.plot(kind = 'bar')
```

```python
df.plot(kind = 'bar')
```

```python
df.plot(kind = 'bar', stacked = True)
```

```python
df.plot(kind = 'bar', stacked = True, grid = True, title = 'Gráfica ejemplo Pandas')
```

```python
df.Y.value_counts()
```

```python
df.Y.value_counts().plot(kind = 'bar', xlabel = 'Y values', ylabel = 'Frequency', color = 'green')
```
